# Multi-faceted Bar plot


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sort the data by 'work_year' for better visualization
sorted_data = data.sort_values(by=['work_year'])

# Set a larger figure size and create a multi-faceted bar plot
plt.figure(figsize=(14, 10))
facet = sns.catplot(x='employment_type', hue='experience_level', col='work_year', kind='count', data=sorted_data, height=5, aspect=1.5, palette='muted')
facet.set_axis_labels('Employment Type', 'Count')
facet.fig.suptitle('Multi-faceted Bar Plot of Experience Level by Employment Type for Each Year', y=1.02)

plt.show()
```

# Description:

The multi-faceted bar plot illustrates the distribution of 'experience_level' within each 'work_year' for different 'employment_types'.

-The plot is divided into facets based on the 'work_year'. Each facet represents a different year. -Within each column (year), there are bars for each 'employment_type'. The x-axis represents the different employment types, and the bars are colored based on the 'experience_level'. -The height of each bar indicates the count of individuals for a specific 'experience_level' within each 'employment_type' for a given year.

# Employment Type

PT: Part-time
FT: Full-time
CT- Contract based
FL- Freelancer

# Experience Level

EN: Entry level
MI: Mid level SE: Senior level EX: Executive level

According to this dataset, it is learnt that, the highest employment was seen in senior level roles that too in the year 2023,with the lowest being in the year 2020, for Executive level roles. And the dataset only focused on Full-Time employee type roles. (As everything else went null on the graph)

# HeatMap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with columns: work_year, experience_level, salary_in_usd
heatmap_data = data.pivot_table(index='work_year', columns='experience_level', values='salary_in_usd', aggfunc='mean')

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)
plt.title('Heatmap of Average Salary by Work Year and Experience Level')
plt.show()
```

# Description:

This Heatmap provides a visual representation of the average salary values for different combinations of work years and experience levels

-Each column on the x-axis represents a different experience level, such as entry-level, mid-level, senior-level, etc.

-Y-Axis (Work Year): Each row on the y-axis represents a different work year. The heatmap is organized in a grid, where each cell corresponds to a specific combination of work year and experience level.

-Color Intensity: The color intensity in each cell indicates the average salary for the corresponding combination of work year and experience level. Darker colors typically represent higher average salaries, while lighter colors represent lower average salaries.

-Annotations: If annot=True is set, numeric values are displayed within each cell, providing the exact average salary value for that combination.

# Interpretation

The highest avg. salary was seen in the year 2023 for Executive level role while the least was seen for entry level role in the year 2021.



# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data['salary_in_usd'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Salaries in USD')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

# Interpretation

The histogram provides a distribution of salaries in USD. The bars represent the frequency of different salary ranges, and the kernel density estimate (KDE) curve gives an estimate of the probability density of the salary variable. This visualization helps to understand the overall distribution and concentration of salaries.

The most salaries were in the range of 100K-200K (usd).

# BoxPlot

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=data, palette='pastel')
plt.title('Boxplot of Salaries by Experience Level')
plt.show()
```

# Interpretation

The boxplot displays the distribution of salaries for each experience level. It shows the median (line inside the box), quartiles (box), and potential outliers (points outside the whiskers). This plot allows to compare the central tendency and spread of salaries across different experience levels.

-The highest salaries were obviously given to executive level roles.

Median line shows the the central tendency i.e the clustering of salary for that experience level. -The Median lines for SE level is at nearly 140K, shows that most salaries for SE fall in this range -The Median lines for MI level is at nearly 100K, shows that most salaries for MI fall in this range -The Median lines for EN level is at nearly 80K, shows that most salaries for EN fall in this range -The Median lines for EX level is at nearly 195K, shows that most salaries for EX fall in this range
-Lowest salaries were given to Entry level roles.

# PieChart


```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with the 'experience_level' column
# Replace column names accordingly if needed
experience_counts = data['experience_level'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(experience_counts, labels=experience_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title('Distribution of Experience Levels')
plt.show()
```
